```python
import time
import numpy as np
import matplotlib.pyplot as plt
from Ping_pong import znajdz_X
```

```python
matrix_sizes = [10, 50, 100, 250, 500, 750, 1000]
n = 10 
```

Macierze 10x10

```python
times10 = [0 for _ in range(10)]
for j in range(n):
    with open(f"dane_testowe/in10_{j}.txt", "r") as f:
        m = int(f.readline())
        A = []
        for _ in range(m):
            A.append(eval(f.readline()))
    start = time.time()
    znajdz_X(A)
    end = time.time()
    times10[j] = end - start
print(times10)
```
